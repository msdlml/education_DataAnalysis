```python
import pandas as pd
import matplotlib.pyplot as plt
import csv

seoul = r'C:\Users\user\Desktop\my_ws\edu\data_analysis\1025_pandas_mat\seoul.csv'
f = open(seoul,encoding='euc-kr')
data = csv.reader(f)
next(data)
day=[]
for _ in range(31):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))

plt.boxplot(day)
plt.show()

```

```python


f = open('age.csv',encoding='euc-kr')
data = csv.reader(f)
next(data)
age = []
name = input("   : ")
for row in data:
    if  name in row[0]:
        for i in row[3:]:
            age.append(int(i))
print(age)
plt.style.use('ggplot') ## <<   
plt.rc('font', family='Malgun Gothic')
plt.plot(age,'r')
plt.title(name+' ')
plt.show()

```

```python

f = open('gender.csv',encoding='euc-kr')
data = csv.reader(f)
next(data)

mage = []
wage = []
name = input("   : ")
for row in data:
    if  name in row[0]:
        for i in row:
            if '' in row[i] and '' in row:
                mage.append(int(i))
            if '' in row[i] and '' in row:
                wage.append(int(i))
plt.style.use('ggplot') ## <<   
plt.rc('font', family='Malgun Gothic')
plt.plot(mage,'blue',label = '')
plt.plot(wage,'r',label = '')
plt.title(name+' ')
plt.legend(loc='upper right')
plt.show()
```

```python
f = open('gender.csv',encoding='euc-kr')
data = csv.reader(f)
next(data)
mage = []
wage = []
name = input("   : ")
for row in data:
    if  name in row[0]:
        for i in row[3:104]:
            mage.append(-int(i))
        for i in row[106:]:
            wage.append(int(i))
plt.style.use('ggplot') ## <<   
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize = (10,5), dpi=300)
plt.barh(mage,width=1,label = '')
plt.barh(wage,width=1,label = '')
plt.title(name+' ')
plt.legend(loc='upper right')
plt.show()
```
