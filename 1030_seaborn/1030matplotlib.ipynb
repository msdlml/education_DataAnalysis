```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('fine_dust.xlsx',index_col='area') #인덱스로 사용할 열의 이름을 나타낸다.
data2018 = data[2018]
plt.figure(figsize=(15,5))
plt.plot(data2018,color = 'b',marker='o')
plt.xlabel('area') #x축 제목
plt.ylabel('micrometer') #y축 제목
plt.title('2018 fine dust line graph') #그래프 제목
plt.grid() #그래프 안에 사각형 패턴
plt.show()
```

```python
plt.figure(figsize=(15,5))
for i in range(2016,2020):
    chardata = data[i]
    plt.plot(chardata,marker='s',label=i)

plt.xlabel('area') #x축 제목
plt.ylabel('micrometer') #y축 제목
plt.title('2016~2019 fine dust line graph') #그래프 제목
plt.legend()
plt.grid() #그래프 안에 사각형 패턴
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('fine_dust.xlsx',index_col='area') #인덱스로 사용할 열의 이름을 나타낸다.
data2017 = data[2017]
plt.figure(figsize=(15,5))
plt.bar(data2017.index,data2017,color = 'b')
plt.ylim(35,55)
plt.xlabel('area') #x축 제목
plt.ylabel('micrometer') #y축 제목
plt.title('2017 fine dust bar graph') #그래프 제목
plt.grid() #그래프 안에 사각형 패턴
plt.show()
```

```python
import numpy as np
index = np.arange(4)

plt.figure(figsize=(15,4))

data4 = data.loc['Gyeonggi':'Daegu',2016:2019]
for year in range(2016,2020):
    cdata = data4[year]
    plt.bar(index,cdata,width=0.2,label = str(year))
    index = index + 0.2 #그래프 위치 조정

plt.ylim(35,55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.xticks(index - 0.5,['Gyeonggi','Incheon','Busan','Daegu']) #x축 눈금 조정 
plt.title('2016~2019 fine dust group bar graph')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
index = np.arange(4)

data4 = data.loc['Gyeonggi':'Daegu',2016:2019]
for year in range(2016,2020):
    cdata = data4[year]
    plt.barh(index,cdata,label = str(year))


plt.xlim(30,55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.yticks(index ,['Gyeonggi','Incheon','Busan','Daegu']) #y축 눈금 조정 
plt.title('2016~2019 fine dust group bar graph')
plt.legend()
plt.show()
```

```python
data = pd.read_excel('health_screenings_2020_1000ea.xlsx')

hdata = data['height']
wdata = data['weight']
plt.figure(figsize= (10,4))
plt.scatter(hdata,wdata)
plt.xlabel('h')
plt.ylabel('y')
plt.title('2020 health scatter graph')
plt.grid()
plt.show()
```

```python

import matplotlib.pyplot as plt

systolic_data = data['systolic']
diastolic_data = data['diastolic']
blood_sugar_data = data['blood_sugar']

plt.figure(figsize=(10,6))

plt.scatter(systolic_data, diastolic_data, color='r', edgecolor='w', label='systolicdiastolic')
plt.scatter(systolic_data, blood_sugar_data, color='g', edgecolor='w', label='systolicblood_sugar')
plt.scatter(diastolic_data, blood_sugar_data, color='b', edgecolor='w', label='diastolic*blood_sugar')

plt.xlim(40,180)
plt.ylim(30,300)
plt.title('2020 Health Screenings Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
mdata = data.loc[data.gender == 1,['gender','height']]
fdata = data.loc[data.gender == 2,['gender','height']]

plt.figure(figsize= (10,6))
plt.hist(mdata['height'],bins=7,alpha = 0.5, label='male') #bins 는 그래프간의 간격이다 높을수록 벌어짐
plt.hist(fdata['height'],bins=7,alpha = 0.5, label='female') # alpha 는 투명도

plt.xlim(130,195)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020health')
plt.legend() #범례 
plt.grid()
plt.show()
```

```python
import numpy as np
height = np.array([mdata['height'], fdata['height']], dtype=object)

plt.figure(figsize=(10,6))
plt.hist(height, bins=6, label=['Male','Female'], stacked=True)

plt.xlim(130,195)
plt.xlabel('height')
plt.ylabel('frequency')
plt.title('2020 Health Screenings Male & Female Height Stacked Histogram')
plt.legend()
plt.show()
```

```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('health_screenings_2020_1000ea.xlsx')
data3 = data.loc[:,['city_code','gender','waist']]
male_data = data3[data3['gender']==1].groupby('city_code')['waist'].mean()
female_data = data3[data3['gender']==2].groupby('city_code')['waist'].mean()

plt.figure(figsize=(12, 4))

#이 부분을 해결해보세요.(2줄)

plt.plot(male_data,color = 'b',linestyle = '--',marker = 'o',markersize='10')
plt.plot(female_data,color = 'r',linestyle = '--',marker = 'o',markersize='10')

plt.xlabel('Count')
plt.ylabel('Waist')
plt.title('2020 Health Screenings Waist')
plt.legend()
plt.grid()
plt.show()
```

```python
waistData = data3.loc[100:110,['waist']]

plt.figure(figsize=(10, 5))

plt.bar(waistData.index, waistData['waist'], color='c')

plt.ylim(50, 110)
plt.xlabel('Count')
plt.ylabel('Waist')
plt.title('2020 Health Screenings Waist')
plt.grid()
plt.show()
```
