```python
import pandas as pd

data = pd.read_csv('seoul-metro-2021.logs.csv')

data.info()
```

```python
station_info = pd.read_csv('seoul-metro-station-info.csv')

station_info.info()
```

```python
station_info[station_info['line.num'] == 2]
```

```python
station_sum = data.groupby('station_code').sum()
station_sum
```

```python
station_geo = station_info[['station.code','geo.latitude','geo.longitude']]

```

```python
station_geo = station_geo.set_index('station.code')
station_geo[station_geo.index == 150]
```

```python
morning_data = data[pd.to_datetime(data.timestamp).dt.hour < 9]
morning_sum = morning_data.groupby('station_code').sum()
morning_sum
```

```python
evening_data = data[pd.to_datetime(data.timestamp).dt.hour > 17]
evening_sum = evening_data.groupby('station_code').sum()
evening_sum
```

```python
joindata = station_sum.join(station_geo) ## 이거 오늘 물어보자
join_morning = morning_sum.join(station_geo)
join_morning
join_evening = evening_sum.join(station_geo)
join_evening
```

```python
seoul_in = folium.Map(location=[37.55,126.98],zoom_start=12)
seoul_in
```

```python
 ## 히트맵 플러그인 모듈 탑재
from folium.plugins import HeatMap
HeatMap(data = joindata[['geo.latitude','geo.longitude','people_in']]).add_to(seoul_in)
seoul_in
```

```python
seoul_out = folium.Map(location=[37.55,126.98],zoom_start=12,width=500,height=500)
HeatMap(data = joindata[['geo.latitude','geo.longitude','people_out']]).add_to(seoul_out)
#folium.Marker([latitude, longitude], popup="판교역",tooltip="판교역 입구").add_to(m) << 마커 찍기
#popup > 마우스 클릭시 표기, tooltip >> 마우스 오버시 표기  folium.Icon 을 사용해서 마커 꾸미기 가능
seoul_out
```

```python
lunch = data[(pd.to_datetime(data.timestamp).dt.hour < 15) & (pd.to_datetime(data.timestamp).dt.hour > 10)]
lunch_sum = lunch.groupby('station_code').sum()

joinlunch = lunch_sum.join(station_geo)
lunchin = folium.Map(location=[37.55,126.98],zoom_start=12)
lunchout = folium.Map(location=[37.55,126.98],zoom_start=12)
HeatMap(data = joinlunch[['geo.latitude','geo.longitude','people_in']]).add_to(lunchin)
lunchin
```

```python
HeatMap(data = joinlunch[['geo.latitude','geo.longitude','people_out']]).add_to(lunchout)
lunchout
```

```python
lunchcircle = folium.Map(location=[37.55,126.98],zoom_start=12)

joinlunch250 = joinlunch[joinlunch['people_out']>2500000]
for index, row in joinlunch250[['geo.latitude', 'geo.longitude', 'people_out']].iterrows():    
        folium.Circle(location=[str(row['geo.latitude']), str(row['geo.longitude'])],   color='Red', fill=True,
              radius=(int(row['people_out'])/10000)).add_to(lunchcircle)

lunchcircle
```
