```python
import folium
import pandas as pd
import random
station_all = pd.read_csv(r"seoul-metro-station-info.csv")
linelist = []
colorlist = ['black','red','green','blue','magenta','cyan','yellow','orange','purple']
seoul_1line = folium.Map(location = [37.55, 126.98], zoom_start=12)
for i in range(1,10):
    for index, row in station_all[station_all['line.num'] == i][['geo.latitude','geo.longitude']].iterrows(): 
        if len(linelist) > 2:
            linelist.remove(linelist[0])
        latitude = row['geo.latitude']
        longitude = row['geo.longitude']
        linelist.append((latitude,longitude))
        folium.PolyLine(locations = linelist,color = colorlist[i-1] ,tooltip=f'{i} Line').add_to(seoul_1line)

seoul_1line
```
