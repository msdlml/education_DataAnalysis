```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv',encoding='euc-kr')

csvr = csv.reader(f)
head = next(csvr)
xlist = []
xlist2 = []
for row in csvr:
    if row[2] != '' and row[0].split('-')[1] == '01':
        xlist.append(float(row[2]))
    if row[2] != '' and row[0].split('-')[1] == '08':
        xlist2.append(float(row[2]))

plt.hist(xlist,color='b',bins=100,label='jan')
plt.hist(xlist2,color='r',bins=100,label='Aug')
plt.legend(loc='upper right')
plt.show()
```

```python
f = open('seoul.csv',encoding='euc-kr')

csvr = csv.reader(f)
head = next(csvr)
month = [[],[],[],[],[],[],[],[],[],[],[],[]]
i = 1
for row in csvr:
    if row[-1] != '':
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
#위에 코드 말로 풀어서 설명해보기
plt.boxplot(month)
 ## 그래프에 보이는 원은 이상치 값을 표현한 것 다른 수치에 비해 너무 크거나 작은 값
plt.show()
```

```python
with open("seoul.csv", "r", encoding="cp949") as sel:
    data = csv.reader(sel)
    next(data)

    year = [[],[],[],[],[],[],[],[],[],[],[],[]]
    
    month_list = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
    for row in data:
        for mon in month_list:
            if row[-1] != "":
                month = row[0].split("-")[1]
                if month == mon:
                    year[month_list.index(mon)].append(float(row[-1]))
print(year)
plt.boxplot(year)
plt.show()
```
