```python
import pandas as pd
import matplotlib.pyplot as plt
filename=r'c:\Temp\iris.csv'
df=pd.read_csv(filename)
df.head()
df.sort_values(by='SepalWidth',ascending=False)
```

```python
plt.rc('font', family='Malgun Gothic') 
newdf=df.sort_values(by='SepalWidth',ascending=False)
top10=newdf.head(10)
top10
top10.plot(kind='bar', rot=0)
plt.title("TOP10")
plt.xlabel("Data")
plt.show()
```
