```python
import bs4
import urllib.request
import time
#제목, 링크, 언론사, 출간일
def newsinfo(news_tag):
    sub = news_tag.find('h2',{'class':'tit'}).text
    link = news_tag.find('a')['href']
    dates = news_tag.find('span',{'class':'medium'})
    date = dates.find('em').text
    pub = news_tag.find('span',{'class':'medium'}).text.replace(date,'').replace('\t','').replace('\n','')
    
    return [sub,link,pub,date]
    
siteurl = "https://news.nate.com/recent?mid=n0100"
htmlObject = urllib.request.urlopen(siteurl)
webPage = htmlObject.read()
bsObject = bs4.BeautifulSoup(webPage, 'html.parser')

big = bsObject.find('div',{'class':'postSubjectContent'})
big_list = big.findAll('div',{'class':'mduSubjectList'})
while True:
    for li in big_list:
        print(newsinfo(li))
    time.sleep(30)
```
