```python
import csv
import time
import datetime

csvName =  'datetime.csv'
with open(csvName, 'w',encoding='euc-kr', newline='') as csvFp:
    csvWriter = csv.writer(csvFp)
    csvWriter.writerow(['', ''])

count = 10
while count > 0 :
    count -= 1

    now = datetime.datetime.now()
    print(now)
    yymmdd = now.strftime('%Y-%m-%d')
    hhmmss = now.strftime('%H:%M:%S')
    time_list = [yymmdd, hhmmss]
    print(time_list)

    #with open(csvName, 'a', newline='') as csvFp:
    with open(csvName, 'a', newline='') as csvFp:
        csvWriter = csv.writer(csvFp)
        csvWriter.writerow(time_list)

    time.sleep(3)
```

```python
import csv
import time
import datetime
import urllib.request
import bs4

nateUrl = "https://news.nate.com/weather?areaCode=11D20401"
htmlObject = urllib.request.urlopen(nateUrl)
webPage = htmlObject.read()
bsObject = bs4.BeautifulSoup(webPage, 'html.parser')

csvName =  'weather.csv'
with open(csvName, 'w',encoding='euc-kr', newline='') as csvFp:
    csvWriter = csv.writer(csvFp)
    csvWriter.writerow(['', '','','','',''])
count = 10
while count > 0 :
    count -= 1
    
    tag = bsObject.find('div',{'class':'right_today'})
    weather_list = tag.find('p',{'class':'celsius'})
    wl1 = tag.find('p',{'class':'humidity'})
    wl2 = tag.find('p',{'class':'rainfall'})
    wl3 = tag.find('p',{'class':'wind'})
    now = datetime.datetime.now()
    print(now)
    yymmdd = now.strftime('%Y-%m-%d')
    hhmmss = now.strftime('%H:%M:%S')
    time_list = [yymmdd, hhmmss,weather_list.text,wl1.text,wl2.text,wl3.text]
    print(time_list)

    #with open(csvName, 'a', newline='') as csvFp:
    with open(csvName, 'a',encoding='euc-kr', newline='') as csvFp:
        csvWriter = csv.writer(csvFp)
        csvWriter.writerow(time_list)
        

    time.sleep(10)
    
```

```python
import urllib.request
import bs4
import openpyxl

# 
outWorkbook = openpyxl.Workbook()
outWorkbook.remove(outWorkbook['Sheet'])
outSheet = outWorkbook.create_sheet('book')


siteurl = "https://www.yes24.com/24/Category/Display/001001046001"
htmlObject = urllib.request.urlopen(siteurl)
webPage = htmlObject.read()
bsObject = bs4.BeautifulSoup(webPage, 'html.parser')

tag_li = bsObject.find('ul',{'class':'clearfix'}) #  
li_list = tag_li.findAll('div',{'class':'goods_info'}) 
rank = 1
header = ['', '', '', '',' ','']
for i in range(1,len(header)+1):
    outCell = outSheet.cell(row=1,column=i,value = header[i-1]) #
for li in li_list: #
    tag_name = li.find('div',{'class':'goods_name'})
    name = tag_name.find('a').text
    tag_auth = li.find('span',{'class':'goods_auth'})
    auth = tag_auth.find('a').text
    pub = li.find('span',{'class':'goods_pub'}).text
    date = li.find('span',{'class':'goods_date'}).text
    price = li.find('em',{'class':'yes_b'}).text+""
    info = [rank,name, auth, pub, date, price]
    for col in range(1,len(info)+1):
        outCell = outSheet.cell(row = li_list.index(li)+2, column= col, value = info[col-1])
    rank += 1
    
outWorkbook.save('bookcrawl.xlsx') 
```

```python
import bs4
import urllib.request

##  
def getBookInfo(book_tag) :
    names = book_tag.find("div", {"class": "goods_name"})
    bookName = names.find("a").text
    auths = book_tag.find("span", {"class": "goods_auth"})
    bookAuth = auths.find("a").text
    bookPub = book_tag.find("span", {"class": "goods_pub"}).text
    bookDate = book_tag.find("span", {"class": "goods_date"}).text
    bookPrice = book_tag.find("em", {"class": "yes_b"}).text
    return [bookName, bookAuth, bookPub, bookDate, bookPrice]

##  
page = 1


##  
while page < 31:
    bookUrl = f"http://www.yes24.com/24/Category/Display/001001046001?ParamSortTp=05&PageNumber={page}"
    htmlObject = urllib.request.urlopen(bookUrl)
    webPage = htmlObject.read()
    bsObject = bs4.BeautifulSoup(webPage, 'html.parser')
    tag = bsObject.find('ul', {'class': 'clearfix'})
    all_books = tag.findAll('div', {'class': 'goods_info'})

    
    for book in all_books :
        print(getBookInfo(book))
    page += 1
    
```
