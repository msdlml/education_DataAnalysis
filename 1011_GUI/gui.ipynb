```python
from tkinter import *
from time import *

flist = ["jeju1.gif","jeju2.gif","jeju3.gif","jeju4.gif","jeju5.gif","jeju6.gif","jeju7.gif","jeju8.gif","jeju9.gif"]
photoList = [None] * 9
num = 0

def clicknext():
    global num
    num += 1
    if num>8:
        num=0
    photo = PhotoImage(file = flist[num])
    pLable.configure(image=photo)
    pLable.image=photo

def clickPrev():
    global num
    num -= 1
    if num<0:
        num=8
    photo = PhotoImage(file = flist[num])
    pLable.configure(image=photo)
    pLable.image=photo

def pageUp(event):
    clicknext()
def pageDown(event):
    clickPrev()

window = Tk()
window.geometry("700x500")
window.title("사진 앨범")

window.bind("<Prior>", pageUp)
window.bind("<Next>", pageDown)

btnP = Button(window, text="<< 이전", command=clickPrev)
btnN = Button(window, text="다음 >>", command=clicknext)

photo = PhotoImage(file=flist[0])
pLable= Label(window, image=photo)

btnP.place(x=250, y =10)
btnN.place(x=400, y =10)
pLable.place(x=15, y =50)

window.mainloop()
```
