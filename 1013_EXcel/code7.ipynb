```python
import xlwt
import csv
import os

csvFilelist = ['singerA.csv','singerB.csv'] #파일이름을 리스트에 저장
outWorkbook = xlwt.Workbook() # 클래스 인스턴스화

for csvFilename in csvFilelist:  
    rowCount = 0  
    with open(csvFilename,'r',encoding= 'euc-kr') as inFp: # 읽기모드로 연다. inFp에 대입
        csvR = csv.reader(inFp) #csvR에 inFp를 리스트화 한 걸 대입
        header_list = next(csvR) #csvR에 첫햇을 헤더리스트에 대입
        outSheet = outWorkbook.add_sheet(os.path.basename(csvFilename)) # csv파일명으로 시트 생성
        for col in range(len(header_list)): #헤더 리스트의 열 갯수만큼 반복
            outSheet.write(rowCount, col, header_list[col]) #엑셀 파일의 0행,col열에 헤더의 col열의 내용을 작성
        for row_list in csvR: #csv파일의 행을 하나씩 호출
            rowCount += 1#카운트를 상승시켜 행바꿈, 반복문 첫번째 위치한 이유는 헤더 스킵
            for col in range(len(row_list)): #열의 갯수만큼 반복
                if row_list[col].isnumeric(): #데이터가 숫자인지 판별
                    outSheet.write(rowCount,col,float(row_list[col])) #숫자일 경우 float형태로 입력
                else: #숫자가 아닐경우
                    outSheet.write(rowCount,col,row_list[col]) #그대로 입력

outWorkbook.save('singerCSV.xls')
```

```python
import csv
import xlrd
workbook = xlrd.open_workbook('singer.xls')
findcol = ['이름','주소','유튜브 조회수']
findidx = []
wsheetList = workbook.sheets()
worksheet = wsheetList[0]
header_list = worksheet.row_values(0)
for name in findcol :
    findidx.append(header_list.index(name))

wsheetList = workbook.sheets()
with open("singer_youtube.csv",'w',encoding='euc-kr', newline='') as outFp:
    csvWriter = csv.writer(outFp)
    csvWriter.writerow(findcol)
    for worksheet in wsheetList :
        for row in range(1, worksheet.nrows) :
            findList = []
            for col in range(worksheet.ncols) :
                if col in findidx :
                    findList.append(worksheet.row_values(row)[col])
            csvWriter.writerow(findList)

print("Save. OK~")
```

```python
import openpyxl
from copy import copy

workbook = openpyxl.load_workbook('singer.xlsx')
wsheetList = workbook.sheetnames

outWorkbook = openpyxl.Workbook()
outWorkbook.remove(outWorkbook['Sheet'])

for wsName in wsheetList:
    worksheet = workbook[wsName]
    outSheet = outWorkbook.create_sheet(wsName)
    for row in range(1, worksheet.max_row+1):
        outSheet.row_dimensions[row].height = worksheet.row_dimensions[row].height
        print(worksheet.row_dimensions[row].height)
        for col in range(1, worksheet.max_column+1):
            outSheet.column_dimensions[chr(ord('A')+col - 1)].width \
                = worksheet.column_dimensions[chr(ord('A')+col -1)].width #ord >> 해당 문자에 해당하는 유니코드 정수를 반환
            inCell = worksheet.cell(row=row,column=col)
            print(inCell)
            outCell = outSheet.cell(row=row, column=col, value = inCell.value)
            print(outCell)
            if inCell.has_style:
                outCell.font = copy(inCell.font)
                outCell.border = copy(inCell.border)
                outCell.fill = copy(inCell.fill)
                outCell.number_format = copy(inCell.number_format)
                outCell.alignment = copy(inCell.alignment)

outWorkbook.save('singer_copy1.xlsx')
print('save')
```

```python
import openpyxl
from copy import copy

workbook = openpyxl.load_workbook('singer.xlsx')
wsheetList = workbook.sheetnames

outworkbook = openpyxl.Workbook()
outworkbook.remove(outworkbook['Sheet'])

for wsName in wsheetList:
    worksheet = workbook[wsName]
    outSheet = outworkbook.create_sheet("New_" + wsName)
    outrow = 0
    for row in range(1, worksheet.max_row+1):
        if row != 1 and int(worksheet.cell(row = row, column =5).value) < 165:
            continue
        outrow += 1
        outSheet.row_dimensions[row].height = worksheet.row_dimensions[row].height
        for col in range(1, worksheet.max_column+1):
            outSheet.column_dimensions[chr(ord('A')+col - 1)].width \
                = worksheet.column_dimensions[chr(ord('A')+col -1)].width
            inCell = worksheet.cell(row=row, column=col)           
            outCell = outSheet.cell(row=outrow ,column=col, value = inCell.value)
            if inCell.has_style:
                outCell.font = copy(inCell.font)
                outCell.border = copy(inCell.border)
                outCell.fill = copy(inCell.fill)
                outCell.number_format = copy(inCell.number_format)
                outCell.alignment = copy(inCell.alignment)

outworkbook.save('singer_copy2.xlsx')


```
